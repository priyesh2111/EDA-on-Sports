# Import Data and required libraries
### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("matches.csv")
data.head()
```

```python
data1 = pd.read_csv("deliveries.csv")
data1.head()
```

```python
season_data=data[['id','season','winner']]
complete_data=data1.merge(season_data,how='inner',left_on='match_id',right_on='id')
```

```python
data.columns.values
```

```python
data = data.drop(columns=["umpire3"],axis=1)
data.head()
```

```python
wins_per_season = data.groupby("season")["winner"].value_counts()
wins_per_season
```

```python
plt.figure(figsize = (18,10))
sns.countplot(x= 'season',data = data, palette="mako")
plt.title("Number of Matches played in each IPL season",fontsize=20)
plt.xlabel("season",fontsize=15)
plt.ylabel('Matches',fontsize=15)
plt.show()
```

### Conclusion : Maximum number of matches were played in 2013,                     2012 and 2011 respectively

```python
plt.figure(figsize= (18,10))
sns.countplot(x='winner', data=data, palette='rocket')
plt.title("Number of matches won by a team", fontsize=25)
plt.xticks(rotation=50)
plt.xlabel("Teams",fontsize=15)
plt.ylabel("No. of wins",fontsize=15)
plt.show()
```

### Conclusion : Mumbai Indians & Chennai Super Kings are the teams with most no. of wins

```python
data['win_by']=np.where(data['win_by_runs']>0,'Bat first', 'Bowl first')
data['win_by']
```

```python
Win=data.win_by.value_counts()
Win
```

```python
labels=np.array(Win.index)
sizes = Win.values
colors = ['#ff1a1a' , '#ffcccc']
plt.figure(figsize=(10,8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', 
        shadow=True,startangle=90)
plt.title('Match Results', fontsize=25)
plt.axis('equal')
plt.show()
```

### Conclusion: Teams won more matches when chasing i.e Bowl first

```python
plt.figure(figsize=(18,10))
sns.countplot(x= 'season', hue='win_by', data=data, palette='PuBu')
plt.title("Number of matches won by batting or bowling first")
plt.xlabel("Season", fontsize=15)
plt.ylabel("Matches won", fontsize=15)
plt.show()
```

### Conclusion: Every year except 2010, more matches were won when teams chose to bowl first.

```python
Toss=data.toss_decision.value_counts()
labels=np.array(Toss.index)
sizes=Toss.values
colors=['#ff1a1a', '#ffcccc']
plt.figure(figsize=(10,8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', 
        shadow=True, startangle=90)
plt.title('Toss Result', fontsize=20)
plt.axis('equal')
plt.show()
```

### Conclusion: Teams chose to field more than bat.

```python
plt.figure(figsize = (18,10))
sns.countplot(x='season', hue='toss_decision',data=data,palette='afmhot')
plt.title("Numbers of matches won by Toss result ",fontsize=20)
plt.xticks(rotation=50)
plt.xlabel("Season",fontsize=15, )
plt.ylabel("Count",fontsize=15)
plt.show()
```

### Conclusion: In recent years, no. of matches by field is more than bat.

```python
final_matches=data.drop_duplicates(subset=['season'],keep='last')
final_matches[['season','winner']].reset_index(drop=True).sort_values('season')
```

```python
match = final_matches.win_by.value_counts()
labels=np.array(Toss.index)
sizes=match.values
colors=['#33FFBD','#DBFF33']
plt.figure(figsize=(10,8))
plt.pie(sizes, labels=labels, colors=colors, autopct="%1.1f%%", 
        shadow=True, startangle=90)
plt.title('Match Results',fontsize=20)
plt.axis('equal')
plt.show()
```

### Conclusion: Even in finals, the winning percentage of teams who were field first won more than bat field

```python
Toss=final_matches.toss_decision.value_counts()
labels=np.array(Toss.index)
sizes= Toss.values
colors=['#F60202','#F3FF00']
plt.figure(figsize=(10,8))
plt.pie(sizes,labels=labels, colors=colors, shadow=True, 
        autopct="%1.1f%%", startangle=90)
plt.title('Toss Result',fontsize=20)
plt.axis('equal')
plt.show()
```

### Conclusion: Percentage of Bat first was more than Field in Toss Result

```python
plt.figure(figsize = (18,10))
top_players = data.player_of_match.value_counts()[:10]
fig, ax = plt.subplots()
ax.set_ylim([0,25])
ax.set_ylabel("Count")
ax.set_title("Man of the Match Winner")
top_players.plot.bar()
sns.barplot(x = top_players.index, y = top_players, orient='v', palette="inferno");
plt.show()
```

```python
four_data=complete_data[complete_data['batsman_runs']==4]
four_data.groupby('batting_team')['batsman_runs'].agg([('runs by fours','sum'),('fours','count')])
```

```python
batsman_four=four_data.groupby('batsman')['batsman_runs'].agg([('four','count')]).reset_index().sort_values('four',ascending=0)
ax=batsman_four.iloc[:10,:].plot('batsman','four',kind='bar',color='black')
plt.title("No of fours hit by players", fontsize=20)
plt.xticks(rotation=50)
plt.xlabel("Player Name",fontsize=15)
plt.ylabel("No of fours",fontsize=15)
plt.show()
```

### Conclusion: Shikhar Dhawan has hit most no. of fours

```python
ax=four_data.groupby('season')['batsman_runs'].agg([('four','count')]).reset_index().plot('season','four',kind='bar',color = 'red')
plt.title("Numbers of fours hit in each season ",fontsize=20)
plt.xticks(rotation=50)
plt.xlabel("season",fontsize=15)
plt.ylabel("No of fours",fontsize=15)
plt.show()
```

### Conclusion: 2013 was the year with most no. of fours hit in a season & 2009 was the least

```python
six_data=complete_data[complete_data['batsman_runs']==6]
six_data.groupby('batting_team')['batsman_runs'].agg([('runs by six','sum'),('sixes','count')])
```

```python
batsman_six=six_data.groupby('batsman')['batsman_runs'].agg([('six','count')]).reset_index().sort_values('six',ascending=0)
ax=batsman_six.iloc[:10,:].plot('batsman','six',kind='bar',color='green')
plt.title("No. of six hit by players")
plt.xticks(rotation=50)
plt.xlabel("Player Name",fontsize=15)
plt.ylabel("No. of sixes",fontsize=15)
plt.show()
```

### Conclusion: Chris Gayle has hit most no. sixes with a big margin.

```python
ax=six_data.groupby('season')['batsman_runs'].agg([('six','count')]).reset_index().plot('season','six',kind='bar',color = 'blue')
plt.title("Nos. of sixes hit in eash season")
plt.xticks(rotation=50)
plt.xlabel("season",fontsize=15)
plt.ylabel("No. of sixes",fontsize=15)
plt.show()
```

### Conclusion: Most no. of sixes were hit in the year 2018  & 2009 was the year with least sixes

```python
batsman_score=data1.groupby('batsman')['batsman_runs'].agg(['sum']).reset_index().sort_values('sum',ascending=False).reset_index(drop=True)
batsman_score=batsman_score.rename(columns={'sum':'batsman_runs'})
print("Top 10 Leading Run Scorers in IPL")
batsman_score.iloc[:10,:]
```

```python
no_matches_played= data1[["match_id","player_dismissed"]]
no_matches_played =no_matches_played.groupby("player_dismissed")["match_id"].count().reset_index().sort_values(by="match_id",ascending=False).reset_index(drop=True)
no_matches_played.columns=["batsman","No_of Matches"]
no_matches_played.head(5)
```

```python
plt.figure(figsize=(18,10))
ax=sns.countplot(x='dismissal_kind',data=data1,palette='viridis')
plt.title("Dismissals in IPL",fontsize=20)
plt.xlabel("Dismissals kind",fontsize=15)
plt.ylabel("count",fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

### Conclusion: Most no. of dismissals in IPL were by caught.

```python
wicket_data=data1.dropna(subset=['dismissal_kind'])
wicket_data=wicket_data[~wicket_data['dismissal_kind'].isin(['run out','retired hurt','obstructing the field'])]
```

```python
top_wickets=wicket_data.groupby('bowler')['dismissal_kind'].agg(['count']).reset_index().sort_values('count',ascending=False).reset_index(drop=True)
ax=top_wickets.iloc[:10,:].plot('bowler','count',kind='bar',color='#0000FF')
plt.title('Most no. of wickets by bowlers',fontsize=20)
plt.xticks(rotation=50)
plt.xlabel('Bowler Name',fontsize=15)
plt.ylabel('No. of wickets',fontsize=15)
plt.show()
```

### Conclusion: Top 3 bowlers with most wickets are Malinga, Amit Mishra & Harbhajan Singh respectively
